```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(X_cat)
encoded = encoder.transform(X_cat).toarray()
encoded
```

```python
X = np.concatenate((x_normalized, encoded), axis=1)
```
